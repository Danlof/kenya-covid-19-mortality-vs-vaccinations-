```python
#load the data
my_filepath=open("H:\datascience\seaborn_vis\covid-vaccination-vs-death_ratio.csv")
my_data=pd.read_csv(my_filepath, index_col="date",parse_dates=True )
```

This is the exploratory analysis for the kenyan covid 19 casestudy

```python
#plot the data for kenya covid 19 casestudy 

my_data3=my_data[my_data["country"].isin(["Kenya"])]

#picking columns of date, total vaccinated people and deaths per day 
data_col=my_data3.filter(['date','people_fully_vaccinated','New_deaths'])

#plotting the above data in lineplot 

plt.figure(figsize=(14,6))
sns.lineplot(data=data_col)
plt.title("A plot of people fully vaccinated and deaths per day against time")
```

```python
#plotting a line graph for deaths per day
plt.figure(figsize=(14,6))
plt.title("line plot for new deaths per day")
sns.lineplot(data=my_data3['New_deaths'])
```

```python
#Average deaths per day
ss.mean(my_data3['New_deaths'])
```

```python
#plotting a scatter plot 

plt.figure(figsize=(14,6))
plt.title("A scatter plot of the percentage of vaccinations in the country vs number of new deaths")
sns.regplot(x=my_data3['New_deaths'],y=my_data3['ratio'])
```
